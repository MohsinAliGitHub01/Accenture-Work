```python
category_scores = final_merged.groupby('Category')['Score'].sum()
category_scores
```

```python
top_5_categories = category_scores.nlargest(5)
top_5_categories
```

```python
top_5_data = final_merged[final_merged['Category'].isin(top_5_categories.index)]
top_5_data['Category'].unique()
```
